## Correct answers stats

```kotlin
val correctDF = DataFrame.read("http://localhost:8080/correct-stats")
correctDF.schema()
```

```kotlin
val correctDF2 = correctDF.add("tq") { question.take(3) + Typography.ellipsis.toString() + question.takeLast(25) }
correctDF2
```

```kotlin
correctDF2.plot {
    bars {
        y(tq) {
            axis.name = "Question"
        }
        x(correct)
    }
}
```
